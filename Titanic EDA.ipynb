### EDA of Age

```python
df.Age.value_counts()
```

<li> Age, appears to be a Continous variable

```python
df[df.Age.isnull()]
```

<li> 177 missing values (NaN) from the variable Age

```python
meanAge = df.Age.mean()
```

```python
df.Age = df.Age.fillna(value=meanAge)
```

<li> Handled the missing values by replacing NaN with the average value of Age

```python
df[df.Age.isnull()]
```

<li> No more missing values now

```python
df.Age.describe()
```

<li> Min, Max, Mean and Standard Deviation values of Sex variable are calculated above

```python
df.Age.hist(bins=10).set_title("Histogram of the Sex of passengers")
```

<li> Graphical representation shows multiple aged passengers are in the ship (most of the passengers age is around 30 which is obviously not accurate as we have replaced the missing values with the Average)
